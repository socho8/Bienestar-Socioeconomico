```python
poblacion = pd.read_csv('../Data/poblacion.csv')
poblacion
```

```python
poblacion.isna().sum()
```

```python
poblacion = poblacion[931:].reset_index()
poblacion = poblacion.rename(columns={'Unnamed: 2' : 'Año', 'Population, density and surface area' : 'Pais', 'T02' : 'NumPais',
                                        'Unnamed: 4' : 'Poblacion estimada (2022)',
                                        'Unnamed: 3' : 'Tipo'})
poblacion = poblacion.drop(columns=['Unnamed: 5', 'Unnamed: 6'])
poblacion = poblacion[poblacion['Tipo'] == 'Population mid-year estimates (millions)']
poblacion = poblacion[poblacion['Año'] == '2022']
poblacion = poblacion.reset_index()
poblacion = poblacion.loc[:,['Pais','Poblacion estimada (2022)']]
poblacion['Poblacion estimada (2022)'] = poblacion['Poblacion estimada (2022)'].str.replace(',', '.')
poblacion['Poblacion estimada (2022)'] = poblacion['Poblacion estimada (2022)'].str.replace(r'\.\d+\.', '.', regex=True)
poblacion['Poblacion estimada (2022)'] = poblacion['Poblacion estimada (2022)'].astype(float)
poblacion['Poblacion estimada (2022)'] = poblacion['Poblacion estimada (2022)'].round(2)
poblacion['Poblacion estimada (2022)'] = poblacion['Poblacion estimada (2022)'] * 1000000
poblacion = poblacion.reset_index(drop=True)
poblacion['Poblacion estimada (2022)'] = poblacion['Poblacion estimada (2022)'].astype(int)
```
