```python
#Creamos mapa con la cantidad de poblacion
fig = go.Figure(data=go.Choropleth(
    locations=data['Codigo Pais'],
    z=data['Poblacion'],
    text=data['Pais'],
    colorscale='Viridis',
    autocolorscale=False,
    marker_line_color='white', 
    colorbar_title='Población',
    zmin=0, #Valor minimo de la escala
    zmax=1500000000,  # Valor máximo de la escala
    zmid=750000000,  # Valor medio para la escala
    colorbar_dtick=100000000,  # Incremento de 10 millones en la escala
))
```

```python
#Mostramos el mapa recientemente creado
fig.update_layout(
    title_text='Mapa de Población',
    geo=dict(
        showframe=False,  
        showcoastlines=False,  
        projection_type='equirectangular' 
    )
)
fig.show()
```

```python
#Creamos variables para utilizar en histogramas
top_10 = data.groupby(by = 'Pais')['Poblacion'].sum().sort_values(ascending=False).head(10)
top_10_pbi = data.groupby(by = 'Pais')['PBI per capita (dolares)'].sum().sort_values(ascending=False).head(10)
top_10_idh = data.groupby(by = 'Pais')['IDH'].sum().sort_values(ascending=False).head(10)
top_10_alf = data.groupby(by = 'Pais')['Indice de alfabetizacion'].sum().sort_values(ascending=False).head(10)
top_10_des = data.groupby(by = 'Pais')['Porcentaje de desempleo'].sum().sort_values(ascending=False).head(10)
top_10_hom = data.groupby(by = 'Pais')['Homicidios cada 100 mil habitantes'].sum().sort_values(ascending=False).head(10)
top_10_pob = data.groupby(by = 'Pais')['Indice de pobreza'].sum().sort_values(ascending=False).head(10)
```

```python
#Histograma con el top 10 de paises con mayor poblacion
plt.figure(figsize=(15,8))
sns.barplot(x=top_10.index,
y=top_10.values)
plt.xlabel('Pais')
plt.ylabel('Poblacion')
plt.show();
```

```python
#Histograma con el top 10 de paises con mayor IDH
plt.figure(figsize=(15,8))
sns.barplot(x=top_10_idh.index,
y=top_10_idh.values)
plt.xlabel('Pais')
plt.ylabel('IDH')
plt.show()
```

```python
#Histograma con el top 10 de paises con mayor cantidad de PBI per capita
plt.figure(figsize=(15,8))
sns.barplot(x=top_10_pbi.index,
y=top_10_pbi.values)
plt.xlabel('Pais')
plt.ylabel('PBI per capita (dolares)')
plt.show()
```

```python
#Histograma con el top 10 de paises con mayor indice de alfabetizacion
plt.figure(figsize=(15,8))
sns.barplot(x=top_10_alf.index,
y=top_10_alf.values)
plt.xlabel('Pais')
plt.ylabel('Indice de alfabetizacion')
plt.show()
```

```python
#Histograma con el top 10 de paises con mayor cantidad de desempleo
plt.figure(figsize=(15,8))
sns.barplot(x=top_10_des.index,
y=top_10_des.values)
plt.xlabel('Pais')
plt.ylabel('Porcentaje de desempleo')
plt.show()
```

```python
#Histograma con el top 10 de paises con mayor cantidad de homicidios
plt.figure(figsize=(15,8))
sns.barplot(x=top_10_hom.index,
y=top_10_hom.values)
plt.xlabel('Pais')
plt.ylabel('Homicidios cada 100 mil habitantes')
plt.show()
```

```python
# Creamos la variable de correlacion entre los indices del dataframe
corr = data[['Poblacion','PBI per capita (dolares)', 'IDH', 'Indice de alfabetizacion' ,'Porcentaje de desempleo', 'Homicidios cada 100 mil habitantes', 'Indice de pobreza']].corr()

#Generamos el mapa de calor de la variable creada anteriormente
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Titulo
plt.title('Mapa de calor')
plt.show()
```
